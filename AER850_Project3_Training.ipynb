```python
from ultralytics import YOLO
import yaml

DATA_YAML_PATH = "/content/AER850-Project3/data/data.yaml"
MODEL_PATH = "yolo11n.pt"

model = YOLO(MODEL_PATH)

#Training 1 - 10 epoch, Img 960, batch 2
#Training 2-6 Dont work (a lot of memory issues)
#Training 7 150 Epoch, img 1024, batch 8

# With this setup, ~30s on google collab, Currently 3.5h usage time, maybe 150 epochs?
# image = 960, batch 2
model.train(
    data=DATA_YAML_PATH,
    epochs=150,
    imgsz=1024,
    batch=8 #Starts running into memory issues above this
)
```
